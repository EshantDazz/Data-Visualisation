# Using distplot and displot (Distribution plot)

```python
sns.displot(df['Age'])
plt.show()
```

Doing the same thing in other ways

```python
sns.displot(df['Age'],kde=True)
plt.show()
```

```python
sns.distplot(df['Age'],kde=False,hist=True,rug=True)
plt.show()
#here I have used distplot, not displot
```

```python
sns.distplot(df['Age'],bins=10)   #it will divide the graph into 10 units
plt.show()
```

```python
sns.distplot(df['Age'],bins=5)   #it will divide the graph into 5 units
plt.show()
```

Inferences 

1) Minimum age is around 28 and maximum age is around 76       (We can see that using the rug function used above)

2) Maximum age is between 40 to 70

3) We don't have people having age 10 or 20

# Using Piechart for categorical data

```python
df.head()
```

```python
df.groupby('Sex').size()
```

```python
df.groupby('Sex').size().plot(kind='pie',autopct='%.2f')
plt.show()
```

Inferences:-

From the data that we have 78.98% of the people are Male and 21.02% of the people are female

```python
df.groupby('RestingECG').size().plot(kind='pie',autopct='%.2f')
plt.show()
```

Inference:-

From the dataset that we have we can see 60.13% of the people are having normal RestingECG, 20.48% of the people are having 
LVH(showing probable or definite left ventricular hypertrophy by Estes' criteria) and 19.39% of the people are having
ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)

```python
df.groupby('HeartDisease').size().plot(kind='pie',autopct='%.3f')
plt.show()
```

Inference:-

From the dataset we can see that 55.338% of the people has chances of having a HeartDisease and the rest 44.66% of the people 
are not having a HeartDisease problem

# Violinplot( Categorical and numerical data)

```python
sns.violinplot(y=df['Sex'],x=df['HeartDisease'])
plt.show()
```

Inference:-

Here we can see in the 0 portion Female(F) is in a bulged shape which means more values in that area and Male(M) has a narrow 
one compared to females, which means females have less chances of getting a HeartDisease compared to males

You can also see in the 1 portion of the x axis Male has a bigger bulged area compared to females which also indicates males
have higher chances of getting a HeartDisease

```python
sns.violinplot(y=df['Age'],x=df['HeartDisease'])
plt.show()
```

In the x axis:

0 represents people not having HeartDisease 

1 represents people having HeartDisease








Inference:-
1) People between the age of 35 and 45 are having lesser chances of having a HeartDisease compared to the ones who are having a 
   heart disease
   
2) Maximum people of around age 54 are not having a HeartDisease and maximum people of around age 58 are having a HeartDisease

3) The median of the people not having a HeartDisease is around 49 and the median of the people having a HeartDisease is around 
   56



```python
sns.violinplot(y=df['Cholesterol'],x=df['HeartDisease'])
plt.show()
```

Inference:-

Poeple having 0 cholesterol there are more people who are having the heart diseases and as we go up people having more cholesterol between 130 to 330 are having less chances of having a heart disease

# Using Jointplot for Correlation

```python
sns.jointplot(x = 'Age' , y = 'MaxHR', data = df, kind = 'hex')
plt.show()
```

Let's use the reg function and see how it works

```python
sns.jointplot(x = 'Age' , y = 'MaxHR', data = df, kind = 'reg')
plt.show()
```

Here the line is flowing through the data so we can say that if "Age" is increasing the "MaxHr" will decrease

```python
sns.jointplot(x = 'Age' , y = 'Age', data = df, kind = 'reg')
plt.show()
```

Totally positive correlation

```python
sns.jointplot(x = 'RestingBP' , y = 'FastingBS', data = df, kind = 'reg')
plt.show()
```

Here the line doesn't flow through both the data so no correlation 

```python
sns.jointplot(x = 'Cholesterol' , y = 'FastingBS', data = df, kind = 'reg')
plt.show()
```

Here also the line does not go through the datas so no correlation

```python
sns.jointplot(x = 'Cholesterol' , y = 'MaxHR', data = df, kind = 'reg')
plt.show()
```

With increase in Cholesterol the MaxHr Increases
